# Checking a Parametric Equation Solution

Shown below are the manual steps to a homework problem, where we solve a system of linear equations and find with a free variable, z = s.  I was curious to see if I could check the correctness of solution in sympy, and sure enough, by plugging in what I got, for any real number of the parameter s, the solution ends up being zero.

![](./parametric_problem.png)

```python
x,y,z,s = sp.symbols("x,y,z,s")
f = -3*x + 10*y - 8*z

s = 100 # Substitute any real number here.
display(f.subs(x, -4*s).subs(y, -2/5 * s).subs(z,s))
```
